*Use the code block below to open `sales.csv` with python. Then create a variable called `content` which is equal to a list of each line of the file*

````python
import matplotlib.pyplot as plt
sales = open ("sales.csv", "r")
content = sales.readlines()
for x in content:
    print(x)
````

*Using the `split` function and a `for` loop calculate the total sales in a month*

*Hint: you will need a seperate variable for each month*

````python
jan = 0
feb = 0
mar = 0
del(content[0])
for x in content:
    splt = x.split(",")
    cost = splt[2]
    date = splt[3]
    slt = date.split("/")
    if slt[0] == "1":
        jan += float(cost)
    elif slt[0] == "2":
        feb += float(cost)
    elif slt[0] == "3":
        mar += float(cost)
print(jan)
print(feb)
print(mar)
````

*Using your calculated total sales create a graph to represent this data and write a written analysis*

````python
month = ["January", "Febuary", "March"]
sales = [jan, feb, mar]

plt.plot(month, sales)

plt.ylim(0, 60000)
plt.ylabel("Sales in USD")
plt.xlabel("Month")
plt.title("Total sales over the months of January, Febuary, and March")
````

The x-axis represents the first three months of the year, while the y-axis depicts the total sales made in those three months respectively. The data can be used to track each months profits, and also eventually be used for tax purposes at the end of the year.

*Using the `split` function, seperate each row of data. Then append the customer's name to a list called `customers`*

````python
customers = []
for x in content:
    splt = x.split(",")
    customers.append(splt[0])
del(customers[0])
print(customers)
````

*Using the `Counter` from the `collections` library, and your list of customers create a graph to represent this data and write a written analysis*

````python
customer_count = Counter(customers)

customer = customer_count.keys()
visits = customer_count.values()

plt.rc('xtick', labelsize=50,) 
plt.rc('ytick', labelsize=50)   
plt.figure(figsize=(100,50))
plt.bar (customer, visits)
plt.xticks(rotation='vertical')

plt.xlabel ("Customer", fontsize = 50)
plt.ylabel ("Number of times they have made a purchase", fontsize = 50)
plt.title ("Number of times each customer has made a purchase from the store", fontsize = 50)
````

The x-axis represents each customer that bought from the store, while the y-axis depicts the amount of times each customer made a purchase. The data within the graph can be used to make estimations of who will make purchases regularly and how many times within a certain amount of time. This will also be used to estimate profits, and make the according financial decisions with the other data from their purchases.
