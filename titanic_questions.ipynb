```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("titanic.csv")
data.head()
```

```python
data['Survived'] = data['Survived'].replace([1], 'Survived')
data['Survived'] = data['Survived'].replace([0], 'Died')
data['Pclass'] = data['Pclass'].replace([1], '1st Class')
data['Pclass'] = data['Pclass'].replace([2], '2nd Class')
data['Pclass'] = data['Pclass'].replace([3], '3rd Class')
survival = data.groupby(['Survived', 'Pclass']).Age.count()
```

```python
plt.style.use('ggplot')
survival.unstack().plot(kind='bar')
plt.ylabel ("Amount of people who died/survived")
plt.xlabel ("Class")
plt.title ("Number of people who died/survived in each class")
```

```python
favg = data.groupby(['Pclass']).Fare.mean()
```

```python
favg.plot(kind='bar')
plt.ylabel ("Amount of money paid")
plt.xlabel ("Class")
plt.title ("Average amount of money paid for each class")
```

```python
paysur = data.groupby(['Survived']).Fare.mean()
paysur.plot(kind='bar')
```
